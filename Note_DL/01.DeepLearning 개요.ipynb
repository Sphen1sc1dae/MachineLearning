## Deep Learning
- **머신러닝의 한 종류**
- 인공신경망은 인간의 뇌신경망의 신호 체계를 본따서 만든 구조
- 이미지 인식, 음성 인식, 자연어 처리 등의 다양한 분야에 활용

### <span style='color:cyan'>Deep Learning 과 Machine Learning 의 차이점</span>
- 가장 큰 차이점은 '특징량(Feature) 추출' 이다
- 머신러닝은 사람(분석가)이 이미지 내부에 있는 특징을 지정하는 방식
- 딥러닝에서는 학습데이터에서 기계가 자동으로 특징을 추출
- Deep Learning 은 스스로 학습을 하기 때문에 Machine Learning 보다 많은 데이터가 필요함
- 즉, 머신러닝까지는 인간이 하나하나 가르쳐야 했지만 딥러닝을 이용하면 기계가 자동으로 특징을 학습할 수 있다

### 신경망(Neural Network)
- 인간의 신경망을 본따 만든 네트워크 구조를 의미
- 입력층, 은닉층, 출력층을 지나 처리가 일어나고 최종적인 결과가 출력된다
- <span style='color: cyan'>**이러한 신경망이 은닉층에 2개 이상 추가되면 깊은 신경망(Deep Neural Network)이라고 하며 이를 활용한 기계학습을 '딥러닝'이라고 부른다**</span>
- 딥러닝은 대량의 데이터를 입력해서 학습한다
- 학습이란 유닛(Unit)의 접속 가중치(Weight, Bias)를 조정하는 것을 의미

#### Deep Learning 이 Machine Learning 보다 우수한 분야
- Image Classification
- Image Object detection
> 발전된 분야 : 자율 주행차
- Sequence data(순서가 있는 데이터) 분석 및 예측
> 발전된 분야 : 번역기(문법에 어순이 있기 때문)

## 용어 정의

#### 손실함수(Loss Functino)
- 출력값(예측값)과 정답의 차이
- 출력값과 정답이 일치할수록 손실함수의 값은 작아진다
- 회귀에는 평균절대오차(MAE)와 평균제곱오차(Mean Squared Error: MSE)
- 분류 문제는 크로스 엔트로피(Cross Entropy)를 사용
- 매개변수(Weight, Bias)를 조절해서 손실함수의 값을 최저로 만드는 과정 = 최적화(Optimization)라 한다.
- 최적화 과정은 Optimizer를 통해 이뤄지며 Optimizer 는 역전파(back Propagation)과정을 수행하여 DL모델의 매개변수를 최적화 한다.

#### 최적화(Optimization)
- 대표적인 방법 : 경사하강법(Batch Gradient Descent)
- 반복적으로 손실함수 차이에 대한 모델 매개변수의 미분값(기울기)을 구한 후, 그 미분값의 반대방향으로 매개변수를 조절해 나가면 결국에는 최저 손실함수에 도달한다

#### 역전파(back propagation)
- Optimizer는 손실함수의 값을 최소화 하기 위해 역전파를 사용해 DL 모델의 모든 매개변수를 변경한다
- 손실함수의 값을 최소화 한다는 것은 정답과 예측값의 차이를 최소화 한다는 것이며, 에러율을 최저로 줄인다는 의미

#### Optimizer
- Batch 경사 하강법
- SGD(Stochastic Gradient Descent)

#### 미니배치(mini batch)
- 배치 경사법은 너무 느리고, 리소스도 많이 사용하며, SGD는 확실히 빠르게 학습하지만 모델이 최적화 안되는 경우가 많다
- 때문에, 두 방법의 절충안인 미니배치 방법을 많이 사용한다
- 전체 데이터를 계산해서 매개변수를 변경하는 대신 정해진 양 만큼을 계산해서 매개변수를 최적화 하는 방법
- 최근 DL 학습에서는 대다수가 미니배치를 사용한다

#### 학습률(Learning Rate)
- Adagrad
- Adam

#### DL의 과대적합(Overfitting)
- 과대적합은 학습시와 검증시의 예측력의 차이가 심한 현상
- DL 모델은 학습시 매개변수가 상당히 많다는 점과 학습 횟수에 제한이 없으며 이 두 특징은 과대적합을 불러 올 수 있다

#### 드롭아웃(Drop Out)
- 드롭아웃은 매개변수 중 일정량을 학습 중간마다 무작위로 사용하지 않는 방법

#### 조기종료(Early Stopping)
- 단순히 학습 반복 횟수(Epoch)만 높이면 학습시간이 길어져서 학습데이터에서만 성능이 좋은 현상 발생
- 어느정도 학습률이 좋으면 사용자가 정한 반복횟수 전에 작업을 종료한다
