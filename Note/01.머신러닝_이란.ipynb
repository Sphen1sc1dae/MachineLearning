# Machine Learning Note

## 머신러닝이란 무엇인가?
- 수많은 데이터를 학습시켜 패턴(규칙)을 찾아내고 이를 기반으로 데이터를 분류(Classfication) 하거나 예측(Predict) ➡️ 일반 S/W는 언제나 똑같은 과정을 거쳐 똑같은 결과 출력.
- 머신러닝 S/W(비결정론적 S/W)는 데이터와 알고리즘에 따라 때로는 다른 결과를 추론해 낸다(추론 : 이미 알고있는 것으로부터 논리적 결론을 도출하는 과정).

## 작업 방법
- 어느 정도 규모가 있는 샘플 데이터를 입력으로 넣어 분석 시킨다.
- 그리고 이때 데이터에서 일정한 규칙을 찾아 낸다.
- 그리고 찾아낸 규칙을 기반으로 다른 데이터를 분류하거나 예측하는 것.

## 특징(Feature)과 규칙(Rule) 찾아내는 법
- 특정 추출 데이터가 어떤 특징을 가지고 있는지 찾고 벡터로 만드는 처리 과정
- 어떤 특징(Feature, Column)을 추출할지는 분석가가 결정. ➡️ 이것이 머신러닝의 중요한 포인트
- 회귀분석(y = ax + b) : 데이터가 연속된 값일 때 이를 선형회귀 라고 부른다. 이때 y를 종속변수(Target), x를 독립변수(Feature) 라고 한다.

## 머신러닝의 종류
#### 지도 학습(Supervised Learning)
: 학습시 데이터와 답을 같이 학습한 후, 다른 데이터의 답을 예측한다.
#### 비지도 학습(Unsupervised Learning)
: 학습시 데이터만 학습한 후, 다른 데이터의 규칙서를 찾는다.
- 클러스터 분석(Cluster Analysis)
- <mark>주성분 분석(Pricipal Component Analysis: PCA)<mark>
#### 강화 학습(Reinforcement Learning)
- 부분적으로 답을 입력한 후, 데이터를 기반으로 최적의 답 도출.

## 머신러닝의 흐름
1. 데이터 수집
2. 데이터 가공 ➡️ 분석
3. 데이터 학습
    - (3-1) 학습 방법 선택
    - (3-2) 매개변수(Hyper Parameter) 조정
    - (3-3) 모델 학습
4. 모델 평가
5. 예측력과 오분류표가 좋은 경우 ➡️ 다음 단계 진행 아니면, (3)번 부터 재실행.
6. 성공

## 머신러닝의 응용 분야
#### 클래스 분류(Classification)
- 특정 데이터의 레이블(Target)을 분류할 수 있다.
- 스팸메일 분류, 필기체 인식, 증권 사기 등
#### **이진 분류**
: 둘 중 하나의 값으로 분류하는 경우.
#### 다중 분류
: 여러개의 분류값 중에서 하나의 값으로 예측하는 경우.
#### 클러스터링(Clustering) : 그룹 나누기
- 값의 유사성을 기반으로 데이터를 여러 그룹으로 세분화.
- 사용자의 취향을 그룹으로 묶어 사용자 취향에 맞는 광고를 제공.
#### 추천(Recommendation)
- 특정 데이터를 기반으로 다른 데이터를 추천.
- 인터넷 서점에서 구매한 책들을 기반으로 다른 책을 추천.
#### 차원 축소(Dimensionality Reduction)
- 데이터의 특성을 유지하면서 데이터의 양을 줄이는 것.
- 데이터를 시각화 하거나 구조를 추출해서 용량을 줄여 계산을 빠르게 하거나 메모리를 절약한다.

#### 머신러닝의 필수 개념
##### 과적합(Overfitting)
- 학습데이터외의 데이터에는 정확도가 낮게 나오는 모델.

##### 과소적합(Underfitting)
- 학습 데이터보다 일반 데이터에 대한 정확도가 더 높게 나오는 모델.
