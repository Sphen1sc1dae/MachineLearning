## 과일분류 문제

```python
import pandas as pd

fruits = pd.read_csv('../Data/fruits.csv')
fruits.head()
```

```python
fruits.info()
```

```python
apple = fruits[:35]
apple.tail()
```

#### apple의 산포도

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(
    apple.length,
    apple.weight
)

plt.xlabel('Length')
plt.ylabel('Weight')
plt.title('APPLE')

plt.show()
```

```python
berry = fruits[35:49]
berry.tail()
```

```python
# berry 산포도
plt.scatter(
    berry.length,
    berry.weight
)

plt.xlabel('Length')
plt.ylabel('Weight')
plt.title('BERRY')

plt.show()
```

```python
# fruits 산포도
plt.scatter(
    fruits.length,
    fruits.weight
)

plt.xlabel('Length')
plt.ylabel('Weight')
plt.title('FRUITS')

plt.show()
```

```python
plt.scatter(
    apple.length,
    apple.weight
)

# berry 산포도
plt.scatter(
    berry.length,
    berry.weight
)

plt.xlabel('Length')
plt.ylabel('Weight')
plt.title('FRUITS')

plt.show()
# Python 에서는 한 셀에 plt 2개 있으면 색으로 구분하여 한 번에 표시
```
