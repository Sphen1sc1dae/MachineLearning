#### RandomForest 이용 버섯 분류

```python
import pandas as pd

mr = pd.read_csv('../Data/mushroom.csv', header=None)
# header = None 옵션 필수. 불러온 데이터의 최상위 행이 컬럼명으로 출력 될 수 있음.
mr.head()
```

- 각각의 행 = 각 버섯의 종류
- 첫번째 열 : p(독), e(식용)
- 두번째, 세번째 열 : 버섯의 머리 모양 > b(벨모양 = 종), c(원뿔), x(볼록), f(평평한), k(혹모양), s(오목)
- 네번째 열 : 버섯의 머리 색깔 > n(갈색), b(황갈색), c(연한갈색)

####  Feature Data => 알파벳 문자 한개 > 숫자로 변환
- ML, DL 에 라이브러리와 함수 사용하기 위해서는 숫자 데이터셋을 사용 하여야 한다.

----
#### 연습

```python
# 알파벳 -> ASCII
ord('x')
```

```python
# ASCII -> 알파벳
chr(120)
```

```python
# mr의 1번컬럼을 숫자로 변경
for i in mr.loc[:,1]:
    print(i, ord(i))
```

```python
target = []
data = []

for row_index, row in mr.iterrows():
    # print(row_index, ":", row)
    # print(row.loc[0])
    target.append(row.loc[0])
    row_data = []
    for v in row.loc[1:]:
        row_data.append(ord(v))
    
    data.append(row_data)
```

```python
print(data[:2])
```

#### Data Frame 생성

```python
# Target 기준
targetTemp = pd.DataFrame(target)
targetTemp.head()
```

```python
# data > dataframe / column 이름 겹치지 않게 생성
dataTemp = pd.DataFrame(data)

dataTemp.rename(columns=lambda x : x+1, inplace=True)

dataTemp.head()
```

```python
# 2개의 DF 합치기
mr2 = pd.concat([targetTemp, dataTemp], axis='columns')
```

```python
mr2.head()
```

```python
mr2.info()
```

##### Random Forest 로 분류

```python
# Train, Test 분리
train_data, test_data, train_target, test_target = \
train_test_split(
    mr2.loc[:,1:],
    mr2.loc[:,0],
    random_state = 42,
    stratify=mr2.loc[:, 0]
)
```

```python
clf = RandomForestClassifier()
clf.fit(train_data, train_target)

print("Training : ", clf.score(train_data, train_target))
print("Test : ", clf.score(test_data, test_target))
```

- 이때 사용하는 것이 One-hot Encoding 이다.
