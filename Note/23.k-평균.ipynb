#### k-Means
- 대표적인 군집 알고리즘. 군집의 중심(평균)을 찾는 작업
- 클러스터(cluster) : 군집
- 클러스터링 : 군집화 하는 작업

```python
import numpy as np
```

```python
fruits = np.load('../Data/fruits.npy')
fruits.shape
```

```python
# 3차원 데이터를 2차원 데이터로 변경
fruits_2d = fruits.reshape(-1, 100*100)
fruits_2d.shape
```

#### k-Means Model

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(
    n_clusters=3
)
km.fit(fruits_2d)
```

```python
# 군집화된 Label 출력
km.labels_
```

---
#### 첫 번째 클러스터 확인하기

```python
import matplotlib.pyplot as plt
```

```python
def draw_fruits(arr, ratio=1): # ratio = 해상도
    n = len(arr)
    # 한 줄에 10개씩 이미지 생성
    rows = int(np.ceil(n/10)) # 올림처리
    
    # 행이 1개라면 열 개수 = 셈플 개수 그렇지 않으면 10개 생성
    cols = n if rows < 2 else 10
    
    fig, axs = plt.subplots(
        rows,
        cols,
        figsize=(cols*ratio, rows*ratio),
        squeeze=False
    )

    for i in range(rows):
        for j in range(cols):
            if i*10 + j < n:
                axs[i, j].imshow(arr[i*10 + j], cmap='gray')
            arr[i,j].axis('off')

    plt.show()
```

```python
draw_fruits(fruits[km.labels_ == 0])
```

```python
draw_fruits(fruits[km.labels_ == 1])
```

#### 클러스터의 중심

```python
draw_fruits(km.cluster_centers_.reshape(-1, 100, 100), ratio=3)
```

---
#### 최적의 k 찾기
##### Elbow Method(최적의 군집수 구하기)

```python
inertia = []
for k in range(2, 7):
    km = KMeans(
        n_clusters=k
    )
    km.fit(fruits_2d) # 불러온 모델로 학습하는 DataSet 은 반드시 2차원 데이터
    inertia.append(km.inertia_)

plt.plot(
    range(2, 7),
    inertia
)

plt.show()
```

> 해당 그래프에서 기울기가 변하는 구간(k = 3 / k = 4)이 최적의 k 값
