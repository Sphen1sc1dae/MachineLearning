#### Color Feature 추가

```python
# Iris-setosa : red | Iris-versicolor : green | Iris-Virginica : blue
color_tmp = []

for name in iris['Name']:
    if name == 'Iris-setosa':
        color_tmp.append('red')
    elif name == 'Iris-versicolor':
        color_tmp.append('green')
    else:
        color_tmp.append('blue')

iris['Color'] = color_tmp

iris.head()
```

#### Machine Learning을 위한 작업

```python
# Feature 만들기
iris_data = iris[
    ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
]
iris_data.head()
```

```python
# Target
iris_target = iris['Name']
iris_target.head()
```

#### 학습용 데이터와 테스트용 데이터로 나누기

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data, train_target, test_target = \
train_test_split(iris_data, iris_target)
```

#### 학습시키기

```python
# 데이터 학습과 예측하기
from sklearn import svm, metrics
```

```python
clf = svm.SVC()
```

```python
clf.fit(train_data, train_target)
```

```python
# 예측하기
pred = clf.predict(test_data)
pred[:5]
```

```python
# 정답률
metrics.accuracy_score(test_target, pred)
```

```python
# 특정 수치를 입력하여 붓꽃 품종 예측 1
clf.predict([
    [4.1, 3.5, 1.4, 0.2]
])
```

```python
# 특정 수치를 입력하여 붓꽃 품종 예측 2
# 학습한 범를 넘어서는 입력값 -> 예측모델 환각 현상
clf.predict([
    [8.1, 9.5, 11.4, 1.2]
])
```
