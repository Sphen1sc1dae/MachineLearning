## 13. 머신러닝 모델

```python
import pandas as pd
```

```python
iris = pd.read_csv("../Data/iris.csv")
iris.head()
```

```python
iris_data = iris[['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']]
iris_label = iris['Name']
```

```python
# Train 과 Test
from sklearn.model_selection import train_test_split
train_data, test_data, train_label, test_label = \
    train_test_split(
        iris_data,
        iris_label,
        random_state=42,
        stratify=iris_label
    )
```

#### kNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier()
knn_clf.fit(train_data, train_label)
print("KNN : ", knn_clf.score(test_data, test_label))
```

#### SVM

```python
from sklearn.svm import SVC
svm_clf = SVC()
svm_clf.fit(train_data, train_label)
print("SVM : ", svm_clf.score(test_data, test_label))
```

#### Gaussian Process

```python
from sklearn.gaussian_process import GaussianProcessClassifier
gp_clf = GaussianProcessClassifier()
gp_clf.fit(train_data, train_label)
print("Gaussian Process : ", gp_clf.score(test_data, test_label))
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier()
dt_clf.fit(train_data, train_label)
print("Decision Tree : ", dt_clf.score(test_data, test_label))
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_clf.fit(train_data, train_label)
print("Random Forest : ", rf_clf.score(test_data, test_label))
```

#### AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
ad_clf = AdaBoostClassifier()
ad_clf.fit(train_data, train_label)
print("AdaBoost : ", ad_clf.score(test_data, test_label))
```

#### Neural Net

```python
from sklearn.neural_network import MLPClassifier
net_clf = MLPClassifier()
net_clf.fit(train_data, train_label)
print("Neural Net : ", net_clf.score(test_data, test_label))
```

#### QDA

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda_clf = QuadraticDiscriminantAnalysis()
qda_clf.fit(train_data, train_label)
print("QDA : ", qda_clf.score(test_data, test_label))
```

#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb_clf = GaussianNB()
gnb_clf.fit(train_data, train_label)
print("Gaussian Naive Bayes : ", gnb_clf.score(test_data, test_label))
```
