## 특성(Feature)공학과 규제(Regularization : 정형화, 일반화, 규제)

```python
import pandas as pd
```

```python
df = pd.read_csv("../Data/fish.csv")
df.head()
```

```python
fish_full = df.to_numpy()
fish_full[:5]
```

```python
df = pd.read_csv("../Data/fish_weight.csv")
df.head()
```

```python
fish_weight = df.to_numpy()
fish_weight[:5]
```

```python
# Train 과 Test
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = \
    train_test_split(
        fish_full,
        fish_weight,
        random_state=42
    )
```

## sklearn 으로 다중회귀 분석 하기

---
#### 다항 특성 만들기 실습

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 다항 특성 만들기
poly = PolynomialFeatures()
poly.fit([[2, 3]])
poly.transform([[2, 3]])
```

```python
poly = PolynomialFeatures(include_bias=False) # degree = 2
poly.fit([[2, 3]])
poly.transform([[2, 3]])
```

```python
poly.get_feature_names_out()
```

---
#### 다항 특성 만들기

```python
train_input.shape
```

```python
# train_input 을 다항 특성 만들기
poly = PolynomialFeatures(include_bias=False) # degree = 2
train_poly = poly.fit_transform(train_input)
```

```python
train_poly.shape
```

```python
# 컬럼 정보 확인
poly.get_feature_names_out()
```

```python
train_poly[:5]
```

```python
# test_input 도 다항특성 만들기
test_poly = poly.transform(test_input)
test_poly.shape
```

#### 다중 회귀 모델 훈련하기

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(train_input, train_target)
print("Train :", lr.score(train_input, train_target))
print("Test :", lr.score(test_input, test_target))
```

```python
lr.fit(train_poly, train_target)
print("Train :", lr.score(train_poly, train_target))
print("Test :", lr.score(test_poly, test_target))
```

---
#### 더 많은 특성 만들기

```python
poly = PolynomialFeatures(include_bias=False, degree=5)
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
```

```python
train_poly.shape
```

```python
poly.get_feature_names_out()
```

```python
lr.fit(train_poly, train_target)
print("Train :", lr.score(train_poly, train_target))
print("Test :", lr.score(test_poly, test_target))
```

> -> 원래 과대적합에 의해 Test 값이 매우 크게 범위에서 벗어나야 함(음수값). 이유는 모르겠지만 안정적인 수로 출력됨

#### 규제
모델이 과적합되게 학습하지 않고 일반성을 가질 수 있도록 규제(정규화) 하는 것

```python
# 규제 전에 표준화
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_poly)

train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

#### Ridge(L2규제) 규제
: 가중치의 제곱으로 규제하는 방법

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()
ridge.fit(train_scaled, train_target)
print("Train :", ridge.score(train_scaled, train_target))
print("Test :", ridge.score(test_scaled, test_target))
```

#### 적절한 규제 강도(Hyper Parameter) 찾기

```python
train_score = []
test_score = []

alpha_list = [0.001, 0.01, 0.1, 1, 10, 100] # 임의로 정한다.

for alpha in alpha_list:
    ridge = Ridge(alpha=alpha)
    ridge.fit(train_scaled, train_target)
    train_score.append(ridge.score(train_scaled, train_target))
    test_score.append(ridge.score(test_scaled, test_target))
```

```python
pd.DataFrame(
    {
        'Train' : train_score,
        'Test' : test_score
    }
)
```

> 2 번째 데이터 즉, alpha 값이 0.1 일 때 가장 좋은 학습 효율

```python
ridge = Ridge(alpha=0.1)
ridge.fit(train_scaled, train_target)
print("Train :", ridge.score(train_scaled, train_target))
print("Test :", ridge.score(test_scaled, test_target))
```

특성공학은 모델 학습 초반 정확도가 너무 안좋을 때 사용할 것
